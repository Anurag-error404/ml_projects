##Importing DataSet and library files

```python
dataset = pd.read_csv('Market_Basket_Optimisation.csv', header = None)
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import pyfpgrowth as fp
```

##Making groups of itemsets in DataSet

```python
transactions = []
for i in range(0, 7501):
  transactions.append([str(dataset.values[i,j]) for j in range(0,20)])

for i in range(0, len(transactions)):
  transactions[i] = [x for x in transactions[i] if str(x) != 'nan']
```

##Setting the value of minumum confidence and minimum support

```python
patterns = fp.find_frequent_patterns(transactions,4)
rules = fp.generate_association_rules(patterns, 0.5)
```

```python
results = list(rules)
```

```python
results
```

##Visualization of output

```python
lhs = rules.keys()
rhs= [rules[i][0] for i in lhs]
confidences = [rules[i][1] for i in lhs]
result = pd.DataFrame({
    'LHS' : lhs,
    'RHS' : rhs,
    'Confidences' : confidences
})
result
```
