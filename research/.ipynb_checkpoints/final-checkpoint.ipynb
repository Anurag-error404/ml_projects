```python
import pandas as pd
import numpy as np
```

```python
#loading data
dfsample = pd.read_csv("mexico_covid19.csv")
```

The dataset has 263,007 instances or records with 41 features, and it contains demographic and clinical data as well as 
results of RT-PCR tests for COVID-19 in patients with a viral respiratory diagnosis. it is filtered down to 10 important
features and a subset of 20000 inctances from the records are chosen for our analysis of different machine learning classifiers.

```python
#loading data
df = dfsample[dfsample.index < 20000]
df.head()
```

The epidemiology dataset of positive and negative COVID-19 cases in Mexico has 41 features/columns written in Spanish. Therefore, the names of the features/columns were translated into English. For the purpose of this work only two (2) demographic features including age and sex and eight (8) clinical features which include pneumonia, diabetes, asthma, hypertension, cardiovascular diseases (CVDs), obesity, chronic kidney diseases (CKDs) and one high-risk factor which is tobacco and RT-PCR test result of COVID-19 in patients dataset were considered

```python
#exploratory data analysis
df.isnull().sum()
#there are no null values
```

```python
#exploratory data analysis
print(df.shape)
print(df.describe())
```

According to the data set description in the research paper, except age, all features of binary type. As evident from the description of our dataframe, we have some wrong values (as maximum value of these features is going till 99). Hence we need to deal with these datapoints.

```python
#checking count of such data
features = ["SEX","PNEUMONIA","DIABETES","ASTHMA","HYPERTENSION","CARDIOVASCULAR","OBESITY", "CKDs", "TOBACCO"]
for f in features:
    count = 0
    for i in df[f]:
        if(i!=0 and i!=1):
            count = count + 1
    print(f, count)
```

```python
#since there are very few such instances, we drop these.
for f in features:
    df = df[df[f] <= 1]
df.describe()
```

```python
#checking count again
for f in features:
    count = 0
    for i in df[f]:
        if(i!=0 and i!=1):
            count = count + 1
    print(f, count)
```

We have removed all invalid data points. We also notice that the age varies from 0 to 102 in our data set. For a practical use case, the mentioned data features will only matter for a person in an average age bracket and will not give meaningful results for someone too old or too young.

```python
#checking the count of too young or too old data points
count = 0
for i in df["AGE"]:
    if(i<=5 or i>=90):
        count = count + 1
print(count)
```

```python
#again since this is not a very big number, we drop such data values
df = df[df["AGE"] < 90]
df = df[df["AGE"] > 5]
df.describe()
```

```python
#now since except the age feature, all other data is binary data, we need to scale the AGE feature
from sklearn import preprocessing

#dividing data into features and target
# create X(Independent data) and y(dependent data)
feature = ["AGE","SEX","PNEUMONIA","DIABETES","ASTHMA","HYPERTENSION","CARDIOVASCULAR","OBESITY", "CKDs", "TOBACCO"]
X = df[feature]
Y = df["RESULT"]
print("X", X.shape)
print("Y", Y.shape)

min_max_scaler = preprocessing.MinMaxScaler()
X_scaled = min_max_scaler.fit_transform(X)
X_scaled[1]
```

The epidemiology dataset for positive and negative COVID-19 cases in Mexico been partitioned into training and testing sets. Therefore, the model were trained the models with 80% training data and tested with the remaining 20% of the dataset.


```python
#train test split
from sklearn import model_selection
X_train, X_test, Y_train, Y_test = model_selection.train_test_split(X_scaled, Y, test_size = 0.2)
print("X_train", X_train.shape)
print("X_test", X_test.shape)
print("Y_train", Y_train.shape)
print("Y_test", Y_test.shape)
#now we should start with implementing models
```

Logistic Regression

```python
#importing
from sklearn.linear_model import LogisticRegression
```

```python
#fitting the data
clf_lr = LogisticRegression(max_iter = 1000)
clf_lr.fit(X_train, Y_train)
```

```python
print("Accuracy score on training data", clf_lr.score(X_train, Y_train))
```

```python
clf_lr_prediction = clf_lr.predict(X_train)
clf_lr_prediction
```

```python
print("Accuracy score on testing data", clf_lr.score(X_test, Y_test))
```

```python
#tweeking some parameters
clf_lr1 = LogisticRegression(max_iter = 1000, C=10000, solver="liblinear")
#reducing the effect of regularisation
clf_lr1.fit(X_train, Y_train)
```

```python
print("Accuracy score on training data", clf_lr1.score(X_train, Y_train))
print("Accuracy score on testing data", clf_lr1.score(X_test, Y_test))
```

```python
from sklearn import metrics
#confusion matrix
prediction_ontrain_lr = clf_lr.predict(X_train)
prediction_ontest_lr = clf_lr.predict(X_test)
print("training data confusion matrix")
print(metrics.confusion_matrix(Y_train, prediction_ontrain_lr))
print("testing data confusion matrix")
print(metrics.confusion_matrix(Y_test, prediction_ontest_lr))
```

```python
import seaborn as sns
cm = metrics.confusion_matrix(Y_train, prediction_ontrain_lr)
sns.heatmap(cm, annot=True, fmt='g')
plt.show()
```
