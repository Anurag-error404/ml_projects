Feature Selection for training the dataset

```python
data1 = data.drop(["District_Name","Crop_Year", "State_Name", "District_Name", "Season", "Crop"],axis=1)
```

```python
data_dum = pd.get_dummies(data1)
data_dum[:5]
```

```python
#Test Train Split
x = data_dum.drop("Production",axis=1)
y = data_dum[["Production"]]
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.33, random_state=42)
print("x_train :",x_train.shape)
print("x_test :",x_test.shape)
print("y_train :",y_train.shape)
print("y_test :",y_test.shape)
```

```python
x_train[:5]
```

```python
# model 1 : linear regression
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train,y_train)
```

```python
preds = model.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
mean_squared_error(y_test,preds)
r2_score(y_test,preds)
```

```python
# model 2 : decision tree
from sklearn.tree import DecisionTreeRegressor

regressor = DecisionTreeRegressor(random_state=42)
regressor.fit(x_train,y_train)
```

```python
preds = regressor.predict(x_test)

mean_squared_error(y_test,preds)
r2_score(y_test,preds)
```

```python
Rice = data[(data["Crop"] == "Rice")]
Rice["production"] = Rice["Production"]/10000
Rice.head()
```

```python
plt.figure(figsize=(14, 10))
sns.regplot(x  = Rice["Area"],y = Rice["production"])
plt.show()
```

```python
plt.figure(figsize=(14, 10))
sns.regplot(y  = Rice["Production"],x = Rice["Crop_Year"])
plt.show()
```

```python
dt = Rice.copy()
```

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, MinMaxScaler
le = LabelEncoder()
scaler = MinMaxScaler()
dt["district"] = le.fit_transform(dt["District_Name"])
dt['season'] = le.fit_transform(dt["Season"])
dt["area"] = scaler.fit_transform(dt[["Area"]])
dt["state"] = le.fit_transform(dt["State_Name"])
```

```python
X = dt[["area", "district", "season", "state"]]
y = dt["Production"]
```

```python
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, BaggingRegressor, AdaBoostRegressor
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVR
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15)
model = RandomForestRegressor()
model.fit(X_train, y_train)
print(model.score(X_test, y_test))
```

Wheat Production

```python
wheat = data[data["Crop"] == "Wheat"]
```

```python
wheat1 = wheat.copy()
wheat1['prod'] = wheat["Production"]/1000
wheat1["area"] = wheat["Area"]/1000
wheat1.dropna(inplace=True)
```

```python
plt.figure(figsize=(14, 10))
sns.regplot(x  = wheat1["area"],y = wheat1["prod"])
plt.show()
```

```python
wheat1["season"] = le.fit_transform(wheat1["Season"])
wheat1["dist"] = le.fit_transform(wheat1["District_Name"])
wheat1["state"] = le.fit_transform(wheat1["State_Name"])
```

```python
X = wheat1[["area", "season", "dist", "state"]]
y = wheat1['prod']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15)
```

```python
model = RandomForestRegressor(n_estimators=100)
model.fit(X_train, y_train)
print(model.score(X_test, y_test))
```
